```python
import pandas as pd
from unidecode import unidecode


# Função remover acentos
def strip_accents(text):
    return str(unidecode(text))

df = pd.read_csv('Relatorios/evento_adverso.csv', sep=';', encoding='utf8', low_memory=False)
df_obitos = df[df['OBITO'].str.contains('sim|provavel')]
df_obitos = df_obitos.drop_duplicates(['NUMERO DA NOTIFICACAO'])

# Colunas em Maiúsculo, sem acentos e sem espaços nas bordas
df_obitos.columns = [strip_accents(x.upper()).strip() for x in list(df_obitos.columns.values)]

# Tratamento dos dados
df_obitos = df_obitos.drop_duplicates()
print(f'Banco após drop duplicates: {df_obitos.shape}')
# Obitos do boletim anterior (196)
```

```python
# Ler arquivo de obitos do boletim anterior (be05 = 196)
obitos_be05 = pd.read_excel('obitos/obitos_be05_conferido_final.xlsx')
# transformar os numero de notificaçao numa lista
lista_obitos_be05 = obitos_be05['NUMERO DA NOTIFICACAO'].astype(str).to_list()

# Cria coluna obitos_be05
df['obitos_be05'] = 'nao'

# Pesquisa lista_obitos_be05 no df evento_adverso e adiciona sim na coluna nova
df.loc[df['NUMERO DA NOTIFICACAO'].astype(str).str.contains('(?i)' + '|'.join(lista_obitos_be05), na=False), 'obitos_be05'] = 'sim' 
```

```python
# Verificar óbitos do be anterior e do atual e criar um arquivo para conferencia
filtro = (df['obitos_be05'] == 'sim') + (df['OBITO'] == 'sim')
df_obitos = df[filtro]
```
